#### Data Preprocessing 진행

* `column : workclass`에 존재하는 `?`의 값은 의미가 없을 것 같습니다.<br/> 애초에 모르는 값으로 데이터가 수집되었기 때문입니다. <br/> 
`?`값을 변경합니다. => `np.nan` 

* `target : salary` 값을 범주형으로 변경합니다

```python
# ? 값이 존재하는 컬럼 'workcalss' 의 고유값 분포는 다음과 같습니다.
print(data['workclass'].value_counts())  # '?' 개수 : 1836개가 빠질 예정입니다..

# 데이터를 날려먹기 전의 shape은 다음과 같습니다. 
print(f'? 값을 지우기 전의 데이터 shape {data.shape}')

# ?값을 drop 하여 봅시다.
data = data.replace('?', np.nan).dropna().reset_index(drop = True)

# 변경된 값을 확인합니다. 
print(f'? 값을 지우고 난 이후의 데이터 shape {data.shape}')

# 타겟값의 형태를 살펴봅니다.
print('salary befor preprocessing', data['salary'].unique())

# 타겟값을 변환합니다. (false & True)
data['salary'] = data['salary'].apply(lambda x:x == '>50K')

# 타겟의 고유값을 확인합니다.
print('salary after preprocessing', data['salary'].unique())
```

```python
# 필요한 라이브러리를 가져옵니다.
from sklearn.model_selection import train_test_split

# 학습데이터 및 검증 데이터셋을 분리합니다.
train, val = train_test_split(data, test_size=0.2, random_state = 42)

# 사용할 특성 및 타겟을 선택합니다.

# 명목형 변수는 다음과 같습니다.
cat_features = [    
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]

# 수치형 변수는 다음과 같습니다.
num_features =  ["age", "capital-gain", "capital-loss", "hours-per-week"]
target = 'salary'

# 특성과 타겟을 분리합니다.
# 훈련
X_train = train[cat_features + num_features]
y_train = train[target]

# 검증
X_val = val[cat_features + num_features]
y_val = val[target]
```

#### XGBoost 라이브러리를 사용하여 Gradient Boosting Decision Tree Classifier로 데이터를 모델링합니다.
* GBDT는 DT의 특성을 그대로 지니기 때문에, `OrdinalEncoding`을 사용합니다.
  * `OneHotEncoding`을 사용하지 않습니다!
* early_stopping 기능을 사용하여 50회까지의 검증데이터셋의 성능향상이 없다면 종료합니다.

```python
# 필요한 라이브러리를 불러옵니다. 
from category_encoders import OrdinalEncoder
from sklearn.metrics import classification_report, roc_auc_score
from xgboost import XGBClassifier

# 인코더를 생성합니다. : 인스턴스 객체
encoder = OrdinalEncoder()

# 훈련데이터 fit_transform
X_train_encoded = encoder.fit_transform(X_train)

# 검증데이터 transform
X_val_encoded = encoder.transform(X_val) 

# boosting 생성
boosting = XGBClassifier(
    n_estimators= 1000, # 트리 모델의 개수 
    objective = 'binary:logistic',# : 이진분류를 위한 로지스틱 회귀, 클래스가 아닌 예측 확률 반환
    learning_rate = 0.2, #: 학습 단계별로 이전 결과를 얼마나 반영할지 설정한다. 일반적으로 0.01 ~ 0.2 사이의 값을 쓰며 너무 작게 하면 시간이 더 오래 걸리게 된다.
    subsample =0.8, # 각 트리마다 데이터 샘플링 비율 over-fitting을 방지한다. 일반적으로 0.5 ~ 1을 사용.
    colsample_bytree = 0.8 # 각 트리마다 feature 샘플링에 대한 비율
    )


eval_set = [
    # 훈련데이터
    (X_train_encoded, y_train), 
    
    # 검증데이터
    (X_val_encoded, y_val)]

# model - fit.
boosting.fit(X_train_encoded, y_train, 
             eval_set = eval_set,
             early_stopping_rounds = 50)
```

#### 검증데이터에 대한 성능 확인하기
boosting으로 돌린 값들 중 최적의 성능을 보여준 값을 이용합니다.


```python
y_pred_proba = boosting.predict_proba(
    X_val_encoded, 
    iteration_range = (0, boosting.best_iteration))[:,1]


y_pred = boosting.predict(X_val_encoded, iteration_range = (0, boosting.best_iteration))
print('검증 accuracy', boosting.score(X_val_encoded, y_val))
print('검증 AUC', roc_auc_score(y_val, y_pred_proba))
print(classification_report(y_pred,y_pred))
```

# **ICE plot**
> **특정관측치(데이터포인트)**에서의 **특정 특성**의 변화에 따른 모델의 예측 양상변화를 확인할 수 있습니다.

* 검증 데이터셋의 세 번째 데이터에 대해 분석해 봅시다.(4번째 행)

```python
X_val.iloc[[3]]
```

* 위와 같은 인구 통계 정보를 갖는 사람에 대하여, 나이가 바뀔수록 모델이 어떻게 예측을 달리하는지 확인하여 봅시다.

```python
# 예측 양상의 변화를 확인하고자 하는 데이터를 가져옵니다. 이 데이터는 하나의 행입니다. 
one_datapoint = X_val_encoded.iloc[[3]].copy()


# 세 번째 데이터의 age값을 데이터셋에 존재하는 최소나이에서 최대 나이까지의 age 값으로 바꿔볼 것입니다.
# age 범위 생성 : range()로 지정하였으므로 최대값에 1을 더한다, 

age_test_range = range(X_val_encoded['age'].min(), X_val_encoded['age'].max()+1)


# 값을 담을 빈 리스트 생성.
results = []

global one_datapoint_pred_proba

for age in age_test_range:
  # age 값을 바꿔 넣어 봅니다.
  one_datapoint['age'] = age  #위에서 뽑아놓은 데이터에서 'age'의 값만 변화합니다. 
  one_datapoint_pred_proba = boosting.predict_proba(one_datapoint)[:,1] # 모델이 age가 바뀐 데이터로부터 확률을 예측합니다.
  print(one_datapoint_pred_proba)
  results.append(one_datapoint_pred_proba.item())
```

```python
results = np.array(results)
results -= results[0]  # 나이가 최소일 때 모델 예측 확률로부터의 변화값을 볼 것입니다.
                       # 최소값의 나이를 뺀다면, 나이가 높아졌을 때 얼마나 값이 변화하는지 알아볼 수 있습니다. 

results
```

```python
# matplotlib으로 시각화를 진행하여 봅시다.
import matplotlib.pyplot as plt

plt.plot(age_test_range, results)
plt.xlabel('age')
plt.ylabel("$\Delta$ pred_proba")
```

#### 나이를 변경했을 때 시각화 결과는 다음과 같습니다.
* 위 정보를 갖는 특성들에 대하여, 모델은 나이가 60에 이르기까지 비교적 꾸준히 연봉이 증가할 것이라고 분석합니다. ($50K를 초과할 확률)

❗ 다른 특성을 갖는 사람들에 대해서도 모델이 위와 같이 분석하는지 확인하여 봅시다.

```python
# 함수를 생성합니다.
# 데이터셋(data), 분석할 데이터의 index(data_index), 변화를 확인할 특성명(target_feature), 바꿔넣어 볼 특성의 범위(target_feature_range)를 입력할 시,
# 해당 범위에서 모델의 예측 양상의 변화를 반환하는 함수를 작성합니다.


def get_ice_plot_data(data, data_index, target_feature, target_feature_range):
  
  # data_index의 값을 통해 하나의 행을 골라줍니다.
  one_datapoint = data.iloc[[data_index]].copy()

  # 결과를 담을 빈 리스트를 생성합니다.
  results = []

  # 타겟 특정 피처의 값의 범위를 지정하여 for문으로 살펴봅니다.
  for target_value in target_feature_range:

    # 지정한 값의 범위는, 지정한 피처(특성)의 값이 됩니다, for문을 통해 변화하는 값을 확인할 수 있습니다.
    one_datapoint[target_feature] = target_value

    # 위에서 생성한 XGBoost 모델을 통해, 우리가 지정한 데이터(one_datapoint)의 예측값을 생성합니다. 
    one_datapoint_pred_proba = boosting.predict_proba(one_datapoint)[:,1]
    
    # 예측값은 현재 array 상태입니다, 하나의 스칼라로 변경합니다. #item() 함수 이용. 
    results.append(one_datapoint_pred_proba.item())
  
  #array 배열을 생성합니다.
  results = np.array(results)
  

  # 배열에서 최소값을 뺀 값을 확인합니다.
  # 최소값을 제거한다면, 각 값의 변화율을 확인할 수 있습니다. : results - results[0]
  # target_feature_range : 타겟의 범위
  return target_feature_range, results - results[0]
```

✅ 0번, 10번, 129번, 2022번 데이터에 대해 동일한 작업을 수행하고 시각화합니다.
* 데이터의 행을 의미합니다.

```python
#  타겟 피처를 지정합니다. : age
target_feature = 'age'
target_feature_range = range(X_val_encoded['age'].min(), X_val_encoded['age'].max()+1)

for data_index in [0, 10, 129, 2022]:
  plt.plot(
      *get_ice_plot_data(
          X_val_encoded, data_index, target_feature, target_feature_range
      ),
      label = f'data_idx = {data_index}'
  )

plt.xlabel(target_feature)
plt.ylabel('$\Delta$ pred_proba')
plt.legend()
```

# **Partial Dependence Plot (PDP)**
* ICE plot을 사용하면, 특정 데이터에 대해 특성의 변화에 따른 모델의 예측 양상 변화를 살펴볼 수 있습니다.

* 하지만, 하나의 ICE Plot을 그려본 후, 모델이 해당 특성에 대해 전반적으로 어떻게 분석하고 이해하는지 판단할 수 없습니다. 
 * 위 예시에서도 확인한 것처럼, 다른 인구 통계학적 특성을 가진 사람들에게 모델은 `column : age`의 영향력을 다르게 평가합니다. 

* 우리는 모델의 해당 특성에 대한 전반적인 이해 방식 또한 살펴보고 싶습니다.




####사람들은 각기 다른 특성들을 가지고 있으나, **평균적으로** 모델의 나이의 변화에 따라 연수입을 어떻게 예측하는지 궁금합니다.

* 여러 세세한 조건에서 모델의 예측 변화를 확인하기는 어렵지만, PDP를 통해 전반적인 모델의 해당 특성에 대한 이해 방식을 확인할 수있습니다.
* **PDP는 입력 데이터들에 대한 ICE plot의 평균으로 구할 수 있습니다.**

#### 우리가 작성한 함수를 이용, 검증데이터셋의 데이터에 대하여 PDP(평균-ICE plot)를 그려봅시다.

* [참고] 시간이 조금 걸리는 작업입니다.

```python
from tqdm import tqdm

target_feature = 'age'
target_feature_range = [20, 30, 40, 50 , 60 ,70, 80]

# 값을 저장하는 빈 리스트 생성
all_ice_plot_data = []

# 모든 검증 데이터에 대하여 우리가 작성한 함수로 ICE plot 데이터를 계산합니다.
for data_index in tqdm(range(len(X_val_encoded))):
    all_ice_plot_data.append(
        get_ice_plot_data(
            X_val_encoded, data_index, target_feature, target_feature_range
        )[1]
    )
```

```python
all_ice_plot_data
```

```python
all_ice_plot_data = np.array(all_ice_plot_data)
print(f'all_ice_plot_data.shape : {all_ice_plot_data.shape}')

# 6033개의 ICE plot data를 0번째(data index) 축으로 평균을 내어봅니다.
pdp_data = all_ice_plot_data.mean(0)
pdp_data

print(f'pdp_data.shape : {pdp_data.shape}')

# PDP(평균 ICE plot) data를 시각화 합니다.
plt.plot(target_feature_range, pdp_data)

# label 정리
plt.xlabel(target_feature)
plt.ylabel('AVG : ($\Delta$ pred_proba)')
plt.show()
```

**INSIGHT : 모델 시각화 결과**
* 평균적으로 20대 초반에 비해 **50대** 에서 **연수입이 $50K를 초과할 확률**이 **24% 가량 상승**한다고 **예측** 합니다.

#### PDP plot의 장점.
* 특정 특성에 대한 모델의 복잡한 반응 양상을 이해하기 쉽게 시각화 해준다는 장점이 존재합니다. 


#### PDP plot에 대하여 해석시 주의사항.

**1. PDP는 각 특성간의 독립성을 전제로 합니다.**

* 위 예시에서는 서로 다른 workclass, education 등의 값을 갖는 샘플 집단에 비하여 age 가 동일하게 분포되어있다고 가정합니다.

* 하지만 만약 서로 강한 상관관계가 있는 경우에는 비현실적인 데이터(전제)로 값을 예측하게 됩니다. 
 * ex) 예를 들자면, 17세에 대학원 졸업 학력을 갖는 사람은 거의 존재하지 않습니다.<br/>
 PDP 계산과정에서는 해당 데이터를 모델에게 예측 입력으로 제공하게 됩니다.
 
* 또한, 해석에 주의할 필요가 있습니다.<br/>
  ex) : 높은 수준의 교육수준을 갖는 집단의 나이가 타 집단에 비하여 큰 값을 가질 때. <br/>
    * 나이에 대한 PDP 곡선이 나이 '특성에 의한 것' 인지 교육 수준의 차이에 의한 것인지 구별할 수 없습니다.



**2. PDP를 해석할 때, 살펴보고자 하는 특성값의 분포를 주의 깊게 살펴보아야 합니다.**

* PDP는 실제 해당 특성값을 갖는 데이터가 존재하거나 또는 그렇지 않아도 특성 값을 넣어서 모델의 예측값을 확인합니다.
  * EX) PDP는 학습 데이터나 평가 데이터에 나이가 60~70세 사이인 데이터가 존재하는지 여부와 상관없이 해당 값을 대입하여 모델의 예측값 변화를 살펴봅니다.
  * 이때, 모델이 학습과정에서 해당 특성값을 갖는 데이터를 학습하지 못했다면, 모델은 해당 구간에서 이상한 예측값을 낼 수 있습니다. 
  * 따라서, 실제 데이터 분포에서 특정 값 지점을 갖는 데이터, 구간이 존재하지 않는 데이터가 많이 없다면, 해당 구간에서 PDP의 해석은 조심스럽게 접근해야 합니다. 

# **`pdpbox` 라이브러리 이용하기**

    * `equal` : 동일한 간격으로 특성값 지정을 선택할 수 있습니다. 
  * `num_grid_points` : 몇 개의 특성값 지정을 선택하여 모델의 변화양상을 확인해볼 것인지 선택할 수 있습니다.
