#### 1-3. 특성 엔지니어링 진행.
> 참고 : 여러 아이디어가 생각난다면, 한 번 시도해볼 것.

````python
# 사용할 라이브러리 불러오기
import numpy as np
 
def engineer(df):

  # 적용되어야할 df의 모든 속성을 가져옵니다.
  selected_cols = df.select_dtypes(include = ['number', 'object'])
  
  # 고유값의 개수를 설정. 
  labels = selected_cols.nunique()

  # 카디널리티 조정 => 30보다 적은 카디널리티를 갖는 특성만 선택 ('state' 제외)
  # 인덱스 값으로 설정된 변수를 list로 변경 
  selected_features = labels[labels <= 30].index.tolist()
  
  # 이후 데이터 출력
  df = df[selected_features]

  # 새로운 특성 생성할 것. : behavioral 글자가 들어있는 컬럼은 묶어줄 것
  behavioral = [col for col in df.columns if 'behavior' in col]


  # 열끼리 더함
  df['behaviorals'] = df[behavioral].sum(axis = 1)

  # 제거할 컬럼 뽑아내기
  dels = [col for col in df.columns if ('employment' in col or 'seas' in col)]

  # 컬럼 제거
  df.drop(columns=dels, inplace=True)
        
  return df
````

````python
# 특성 엔지니어링 진행, 훈련 및 검증 및 테스트 데이터 모두 적용

train = engineer(train)
val = engineer(val)
test = engineer(test)
````

````python
# 종속변수 제거, 독립변수 생성
features = train.drop(columns = [target]).columns

# 데이터 분리.

# train data
X_train = train[features]
y_train = train[target]

# val data
X_val = val[features]
y_val = val[target]

#Test data
X_test = test[features]
````

---
#### 파이프라인의 유무 확인


````python
# 필요한 라이브러리 및 패키지 선언
from category_encoders import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
````

> 파이프라인이 존재하지 않는 경우는 다음과 같다.</br>

로지스틱 회귀 예시


````python
# 인스턴스 객체 생성 

# 1. 범주형 변수 -> 연속형 변수
enc = OneHotEncoder()

# 2. 결측값 채우기
imp_mean = SimpleImputer()

# 3. 표준화
scaler = StandardScaler()

# 모델 - 로지스틱 진행
model_lr = LogisticRegression(n_jobs=-1)

#-----------훈련데이터-----------------
# 원핫인코딩
X_train_encoded = enc.fit_transform(X_train)

# 이후 결측값 채우기
X_train_imputed = imp_mean.fit_transform(X_train_encoded)

# 이후 표준화 진행
X_train_scaled = scaler.fit_transform(X_train_imputed)

# 로지스틱 회귀 모델 진행
model_lr.fit(X_train_scaled, y_train)

#-----------검증 데이터----------------


# 원핫인코딩
X_val_encoded = enc.transform(X_val)

# 임퓨터 진행
X_val_imputed = imp_mean.fit_transform(X_val_encoded)

# 표준화 진행
X_val_scaled = scaler.fit_transform(X_val_imputed)


# 정확도 출력
print('검증세트', model_lr.score(X_val_scaled, y_val))
````

````python
# 테스트 데이터 진행

# 원핫인코딩
X_test_encoded = enc.transform(X_test)

# 결측값 채우기.
X_test_imputed = imp_mean.transform(X_test_encoded)

# 표준화진행
X_test_scaled = scaler.transform(X_test_imputed)

# 예측값 출력
y_pred = model_lr.predict(X_test_scaled)
````

#### 파이프라인을 사용하여 모델 생성

````python
pipe = make_pipeline(
    OneHotEncoder(),
    SimpleImputer(),
    StandardScaler(),
    LogisticRegression(n_jobs=-1)
)

pipe.fit(X_train, y_train)
print('검증세트 정확도', pipe.score(X_val, y_val))
y_pred = pipe.predict(X_test)
````

#### 파이프라인에서 모델의 파라미터 등과 같은 정보를 확인하는 방식

````python
pipe.named_steps
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# 로지스틱 접근
model_lr = pipe.named_steps['logisticregression']

# 원핫인코딩 접근
enc = pipe.named_steps['onehotencoder']

# 적용된 검증데이터의 컬럼
encoded_columns = enc.transform(X_val).columns


# 계수 : 기울기값 구하기.
coefficients = pd.Series(model_lr.coef_[0], encoded_columns)

# 시각화
plt.figure(figsize = (10, 30))
coefficients.sort_values().plot.barh()
plt.show()
````

#### 사이킷런 결정트리를 사용하여 결정트리 구현해보기.
> 결정트리에서 표준화는 큰 도움이 되지 않으므로, 제외하고 학습할 것.

````python
from sklearn.tree import DecisionTreeClassifier

pipe = make_pipeline(
    OneHotEncoder(use_cat_names = True),
    SimpleImputer(),
    DecisionTreeClassifier(random_state=1, criterion = 'entropy'))

pipe.fit(X_train, y_train)
print('훈련 정확도', pipe.score(X_train, y_train))
print('검증 정확도', pipe.score(X_val,y_val))
````
